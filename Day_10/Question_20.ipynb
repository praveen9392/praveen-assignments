# 4. plot the daily attendance of two tracks, 'Berri1', 'PierDup'

```python
plt.figure(figsize=(12, 6))
plt.plot(bikes_df['Date'], bikes_df['Berri1'], label='Berri1', color='b',marker="o",markersize=3)
plt.plot(bikes_df['Date'], bikes_df['PierDup'], label='PierDup', color='g',marker="o",markersize=3)
plt.xlabel('Date')
plt.ylabel('Number of Bikes')
plt.title('Daily Attendance Berri1 vs PierDup')
plt.legend()
plt.show()
```

# 5. Check index , explore weekday_name attributes 

```python
bikes_df
```

```python
bikes_df['Date'] = pd.to_datetime(bikes_df['Date'], dayfirst=True)
bikes_df.set_index('Date', inplace=True)
bikes_df['Weekday'] = bikes_df.index.weekday 
bikes_df['Weekday_Name'] = bikes_df.index.day_name() 
```

```python
print(bikes_df[['Weekday', 'Weekday_Name']].head())
```

```python
bikes_df
```

# 6. Get sum of all attendance as a function of the weekday

```python
attendance = bikes_df.groupby(bikes_df['Weekday']).sum()
attendance
```

# 7. Display this in figure , what is the inference?

```python
bikes_df.groupby('Weekday_Name').sum().reindex(
    ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
).plot(kind='bar', figsize=(10, 6), color=['b', 'g'])
plt.xlabel('Day of the Week')
plt.ylabel('Total Bikes')
plt.title('Total Attendance by Weekday')
plt.show()
```

# 20_C. Roger Federer database Each row corresponds to a ATP match played by Roger Federer Can you infer anything on Roger Federer performance?

```python
federer = pd.read_csv('federer.csv')
print(federer.info())
print(federer.head())
```

# 2. How many % of matched won by our player? ('winner')

```python
matches=federer.shape[0]
matches_won=federer[federer['winner']=='Roger Federer'].shape[0]
win_percentage=(matches_won/matches)*100
print(f"Percentage of matches won by Roger Federer: {win_percentage:.2f}%")
matches_won
```

# 3. Proportion of double faults wrt total points in each match 

```python
federer['Double_faults']=federer['player1 double faults']/federer['player1 total points total']
Double_faults=federer['Double_faults']
print(Double_faults)
```

# 4. Average Win per surface 

```python

Avg_wps=federer[federer['winner']=='Roger Federer'].groupby('surface').size()
print(Avg_wps)
```

# 5. Display the proportion of double faults as a function of the tournament date, 'start date' Trend: display average double faults in each year 

```python
federer['start date']=pd.to_datetime(federer['start date'],errors='coerce')
federer['year']=federer['start date'].dt.year
Avg_by_year=federer.groupby('year')['Double_faults'].mean()
print(Avg_by_year)
```

# 20_D. Create two frequencies 5Hz and 50Hz sin signals Draw FFT components 

```python
from scipy.fft import fft, fftfreq

Fs = 500
t = np.linspace(0, 1, Fs)
signal = np.sin(2*np.pi*5*t) + np.sin(2*np.pi*50*t)

fft_values = fft(signal)
freqs = fftfreq(Fs, 1/Fs)
plt.plot(freqs[:Fs//2], np.abs(fft_values[:Fs//2]))
plt.title('FFT of 5Hz + 50Hz Signal')
plt.xlabel('Frequency')
plt.ylabel('Magnitude')
plt.show()
```

# 20_E. Plotting normal random numbers and show that addition of two Gaussian is Gaussian and addition of all is Gaussian

```python
x1 = np.random.normal(0, 1, 10000)
x2 = np.random.normal(2, 1, 10000)
x3 = np.random.normal(5, 2, 10000)

plt.hist(x1 + x2, bins=50, alpha=0.6, label='x1 + x2')
plt.hist(x1 + x2 + x3, bins=50, alpha=0.6, label='x1 + x2 + x3')
plt.legend()
plt.title('Sum of Gaussians')
plt.show()
```
